```python
import pandas as pd
```

```python
df = pd.read_csv('wb_data.csv')
```

```python
df.info()
```

```python
sector_columns = ['sector_vocabulary', 'sector_code', 'sector_percentage', 'iati_identifier']
sector_columns
df_sector = df[sector_columns].dropna(subset=['sector_vocabulary', 'sector_code', 'sector_percentage', 'iati_identifier'])
df_sector
```

```python
df_sector['sector_vocabulary'] = df_sector['sector_vocabulary'].str.split(',')
df_sector['sector_vocabulary']
```

```python
df_sector['sector_code'] = df_sector['sector_code'].str.split(',')
```

```python
df_sector['sector_percentage'] = df_sector['sector_percentage'].str.split(',')
```

```python
df_sector_exploded = df_sector.explode(['sector_vocabulary', 'sector_code', 'sector_percentage'])
df_sector_exploded
```

```python
transaction_columns = ['transaction_value_value_date', 'recipient_country_code', 'iati_identifier', 'transaction_transaction_type_code', 'transaction_value']
transaction_columns
df_transaction = df[transaction_columns].dropna(subset=['transaction_transaction_type_code', 'transaction_value', 'transaction_value_value_date', 'iati_identifier'])
df_transaction
```

```python
df_transaction['transaction_transaction_type_code'] = df_transaction['transaction_transaction_type_code'].str.split(',')
```

```python
df_transaction['transaction_value'] = df_transaction['transaction_value'].str.split(',')
```

```python
df_transaction['transaction_value_value_date'] = df_transaction['transaction_value_value_date'].str.split(',')
```

```python
df_transaction_exploded = df_transaction.explode(['transaction_transaction_type_code', 'transaction_value', 'transaction_value_value_date'])
df_transaction_exploded
```

```python
df_transaction_exploded['transaction_value_value_date'] = pd.to_datetime(df_transaction_exploded['transaction_value_value_date']).dt.year
```

```python
df_exploded_filtered = df_transaction_exploded[df_transaction_exploded['transaction_transaction_type_code'] == '3']
df_exploded_filtered
```

```python
#Merge transaction columns and sector columns

merged_df = pd.merge(df_sector_exploded, df_exploded_filtered, on = "iati_identifier", how = "inner")

merged_df
```

```python
merged_df.info()
```

```python
merged_df['transaction_value'] = merged_df['transaction_value'].str.replace('', '').astype(float)
```

```python
merged_df['sector_percentage'] = merged_df['sector_percentage'].str.replace('', '').astype(float) / 100
```

```python
merged_df_grouped = merged_df.groupby(['iati_identifier', 'recipient_country_code', 'sector_vocabulary', 'sector_code', 'sector_percentage', 'transaction_transaction_type_code', 'transaction_value_value_date'])['transaction_value'].sum().reset_index()
```

```python
merged_df_grouped
```

```python
merged_df_grouped['allocated_amount'] = merged_df_grouped['transaction_value'] * merged_df_grouped['sector_percentage']
merged_df_grouped
```

```python
merged_df_grouped.info()
```

```python
#filter only 2021 and above

WorldBank_data_final_2021to2024 = merged_df_grouped[merged_df_grouped['transaction_value_value_date'] >= 2021]

WorldBank_data_final_2021to2024
```

```python
#save merged_df_grouped as file

merged_df_grouped.to_csv("WorldBank_data_final.csv", index=False)
print("merged_df saved as WorldBank_data_final.csv")
```

```python
WorldBankph = WorldBank_data_final_2021to2024[WorldBank_data_final_2021to2024['recipient_country_code'] == 'PH']
WorldBankph
```

```python
#save as csv

WorldBank_data_final_2021to2024.to_csv("WorldBank_data_final_2021to2024.csv", index=False)
print("WorldBank_data_final_2021to2024.to_csv saved as WorldBank_data_final_2021to2024.csv")
```

```python
WorldBank_data_final_2021to2024PH.to_csv("WorldBank_data_final_2021to2024PH.csv", index=False)
print("WorldBank_data_final_2021to2024PH.to_csv saved as WorldBank_data_final_2021to2024PH.csv")
```
